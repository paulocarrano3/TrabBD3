```python
spark.sql("USE CATALOG trabalho_etl_bd3")
spark.sql("USE SCHEMA bronze")

```

```python
spark.sql("CREATE VOLUME IF NOT EXISTS trabalho_bd3_csv_files")
```

```python
df_clientes = (
    spark.read
    .option("header", True)
    .option("inferSchema", False)
    .csv("/Volumes/trabalho_etl_bd3/bronze/trabalho_bd3_csv_files/clientes.csv")
)
```

```python
df_clientes.show(5)
```

```python
df_pedidos = (
    spark.read
    .option("header", True)
    .option("inferSchema", False)
    .csv("/Volumes/trabalho_etl_bd3/bronze/trabalho_bd3_csv_files/pedidos.csv")
)

df_pedidos.show(5)
```

```python
df_entregas = (
    spark.read
    .option("header", True)
    .option("inferSchema", False)
    .csv("/Volumes/trabalho_etl_bd3/bronze/trabalho_bd3_csv_files/entregas.csv")
)

df_entregas.show(5)
```

```python
df_clientes.write.format("delta").mode("overwrite")\
    .saveAsTable("trabalho_etl_bd3.bronze.clientes")
```

```python
df_pedidos.write.format("delta").mode("overwrite")\
    .saveAsTable("trabalho_etl_bd3.bronze.pedidos")
```

```python
df_entregas.write.format("delta").mode("overwrite")\
    .saveAsTable("trabalho_etl_bd3.bronze.entregas")
```
